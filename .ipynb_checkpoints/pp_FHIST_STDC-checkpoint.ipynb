## 1. Functions

```python
# overview of necessary functions

# plot SEB
def plot_SEB(dSWnet,dLWnet,dSHF,dLHF,dR, threshold):
    ax = plt.subplot(111)

    ind = np.arange(5)  
    ax.bar(ind,[dSWnet,dLWnet,dSHF,dLHF,dR])
    ax.set_xticks(ind)
    ax.set_xticklabels(('$\Delta$SWnet','$\Delta$LWnet','$\Delta$SHF','$\Delta$LHF','$\Delta$R'))
    ax.set_ylabel('Reservoir impact [W m$^{-2}$]')
    ax.set_title('Surface Energy Balance for reservoir cells > '+str(threshold)+'%', loc='right', fontsize=13)
    ax.axhline(0, color='gray', linewidth=1)


def do_STDC(LWout,LWout_ctl, SWin,SWout,SWin_ctl,SWout_ctl,dSWin,dLWin,dLHF,dSHF,dR):
    """Do Surface Temperature Decomposition"""
    
    
    # Stefan Boltzman constant
    sigma = 5.67e-8 # W mÂ² K^-4

    # assumption: emissivity is constant (1)

    # get radiative temperature from LWout (nores) - see equation 4.11 in (Oleson et al., 2010)
    Ts = (LWout/sigma)**(1/4) # LWin for res case -> double check!!
    Ts_ctl = (LWout_ctl/sigma)**(1/4) # LWin for nores case -> double check!!

    dTs = Ts - Ts_ctl
    
    term = 1 / (4 * sigma * Ts**3)

    dT_albedo = term * (-1) * SWin * (SWout/SWin - SWout_ctl/SWin_ctl) # why SWin_res here? 
    dT_SWin   = term * (1 - SWout/SWin) * dSWin
    dT_LWin   = term * dLWin 
    dT_LHF    = term * (-1) * dLHF
    dT_SHF    = term * (-1) * dSHF
    dT_RES    = term * (-1) * dR


    # Wim didn't calculate emissivity -> check why? 
    # also don't include here (suppose is 1, black body)

    dT       = [ dT_LHF, dT_SHF,dT_RES, dT_albedo, dT_SWin, dT_LWin,  dTs]
    
    # Again check closure for dTs (calculated and model diagnostic)

    closure_dT    = dTs - sum(dT[:6]);
    closure_dT_rel = closure_dT / dTs * 100
    
    #print('closure dTs')
    #print(closure_dT)
    #print('closure dT relative')
    #print(closure_dT_rel)
    
    return dT


def plot_STDC(dT_array, threshold, ax = False, ylims = False, panel_label=False):
    """Plot Surface Temperature Decomposition for annual data"""

    if ax == False:
        ax = plt.subplot(111)
        
    cmap = mpl.cm.get_cmap('Paired')
    colors_terms = [cmap(0.2), cmap(0.3), cmap(0.5),cmap(0.4), cmap(0.05), cmap(0.1), 'maroon', 'maroon']

    # plot STDC
    ind = np.arange(8)  
    ax.bar(ind,dT_array, color=colors_terms)

    ax.set_xticks(ind)
    ax.set_xticklabels(('LHF','SHF','R',r'$\alpha$','SWin','LWin','$\Delta$Ts', '$\Delta$T2M'))
    ax.set_ylabel('$\Delta$ Ts forcing [K]')
    ax.set_title('Reservoir grid cells > '+str(threshold)+'%', loc='right');
    ax.set_xlim([ind.min()-1,ind.max()+1])
    ax.axhline(0, color='gray', linewidth=1)
    ax.axvline(5.5, color='lightgray',  linestyle='dashed',linewidth=1)
    if panel_label != False:
        ax.text(0, 1.02, panel_label, color='dimgrey', fontsize=12, transform=ax.transAxes, weight = 'bold') 
    if ylims != False:
        ax.set_ylim(ylims)
    
    
def plot_STDC_seas(dT_month,dT_year, threshold, ax = False, legend = True, ylims = False, panel_label=False):
    """Plot Surface Temperature Decomposition for seasonal cycle and year"""

    if ax == False:
        ax = plt.subplot(111)

    cmap = mpl.cm.get_cmap('Paired')
    colors_terms = [ cmap(0.2), cmap(0.3),cmap(0.5), cmap(0.4), cmap(0.05), cmap(0.1)]
    months = ['J', 'F', 'M', 'A', 'M', 'J', 'J', 'A', 'S', 'O', 'N', 'D']


    # plot monthly STCD
    ind_months = np.arange(12)  
    dT_month = np.asarray(dT_month)
    dT_month_stack = get_bottom(dT_month)

    for i in np.arange(0, dT_month.shape[0]-1):
        ax.bar(ind_months,dT_month[i], bottom=dT_month_stack[i], color=colors_terms[i])

    # plot annual STCD
    ind_year = 12.5
    dT_year = np.asarray(dT_year)
    dT_year_stack = get_bottom(dT_year)

    for j, term in enumerate(dT_year[:-1]):
        ax.bar(ind_year,dT_year[j], bottom = dT_year_stack[j], color=colors_terms[j])

    # plot mean temperature as line
    ax.plot(ind_months,dT_month[-1,:],  '-o', markersize=4, linewidth=1.5, color='maroon', alpha=0.7)
    ax.plot(ind_year,dT_year[-1],  'o', markersize=4, linewidth=1.5, color='maroon', alpha=0.7, label='_nolegend_')

    ind = np.append(ind_months,ind_year)
    # plot settings
    xticklabels = months+ ['year']
    ax.set_xticks(ind)
    ax.set_xticklabels(xticklabels)
    ax.set_ylabel('$\Delta$ Ts forcing [K]')
    ax.set_title('Reservoir grid cells >'+str(threshold)+'%', loc='right');
    ax.set_xlim([ind.min()-1,ind.max()+1]) 
    ax.axhline(0, color='gray', linewidth=1, label='_nolegend_');
    ax.axvline(11.75, color='lightgray',  linestyle='dashed',linewidth=1, label='_nolegend_') ;
    
    # optional arguments
    if legend:
        ax.legend([r'$\Delta$Ts','LHF','SHF','R',r'$\alpha$','SWin','LWin'], loc='center left',  bbox_to_anchor=(1, 0.5), frameon=False)
    if panel_label != False:
        ax.text(0, 1.02, panel_label, color='dimgrey', fontsize=12, transform=ax.transAxes, weight = 'bold') 
    if ylims != False:
        ax.set_ylim(ylims)



    
```
